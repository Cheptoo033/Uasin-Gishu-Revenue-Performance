```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
dp=pd.read_csv('revenue_3(1).csv')
dp
```

```python
x=dp['TEAMS']
y=dp['TOTAL']
```

```python
plt.scatter(x,y,color='black')
plt.title('THIS IS A SCATTERPLOT SHOWING....')
plt.xlabel('TEAMS')
plt.ylabel('TOTALS')

plt.show()
```

```python
plt.hist(y,color='yellow')
plt.title('REVENUE HISTOGRAM')
plt.ylabel('frequency')
plt.xlabel('TOTALS')
plt.show()
```

```python
plt.bar(x,y,color='green')
plt.title('REVENUE BAR GRAPH')
plt.xlabel('TEAMS')
plt.ylabel('TOTALS')
plt.show()
```
